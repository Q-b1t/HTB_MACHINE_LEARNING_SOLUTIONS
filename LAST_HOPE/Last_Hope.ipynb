```python
# receives as parameters a quantum circuit and a backend. It simulates the provided circuit in the given backend the number of times it is specified.
def ideal_qc_simulation(exp,backend,num_measurements = 1024):
  qc_compiled = transpile(exp,backend)
  job = backend.run(qc_compiled,shots = num_measurements)
  result = job.result()
  return result.get_counts()
```

# Download file and instance quatum circuit:


```python
# download content
directory = Path("modules")
filename = "quantum_artifact.qasm"
file_path = directory / filename
raw = "https://raw.githubusercontent.com/Q-b1t/htbml/main/LAST_HOPE/quantum_artifact.qasm"

if not os.path.exists(directory):
  os.makedirs(directory)

if file_path.exists():
  print(f"File {filename} has already beed downloaded.\nSkipping...")
else:
  print(f"Downloading {filename}...")
  request = requests.get(raw)
  with open(file_path,"wb") as f:
    f.write(request.content)
```

```python
# instance the circuit at hand
qc = QuantumCircuit.from_qasm_file(file_path)
```

# Theory:
The result of the quantum measurement is the flag encoded in binary.

```python
len(final_sequence) % 8
```

```python
# separate each letter
binary_encodings = list()
position = 0
step = 8
for i in range(0,len(final_sequence),step):
  binary_encodings.append(final_sequence[position:position+step])
  position += step
```

```python
plain_text = [chr(int(bn,base = 2)) for bn in binary_encodings]
```

```python
plain_text = ''.join(plain_text)
```
